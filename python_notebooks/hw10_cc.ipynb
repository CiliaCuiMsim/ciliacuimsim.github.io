## 1 Rec plot


```python
brush = alt.selection_interval(encodings=['x','y'])
rect = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_bcubcg_fall2022/main/data/building_inventory.csv"},
  "mark":"rect",
  "height":400,
  "width":400,
  "encoding":{
    "x":{"bin":{"maxbins":20},
         "field":"Total Floors", 
         "type":"quantitative"},
    "y":{"field":"Agency Name", 
         "type":"ordinal",
         "axis": {"labelLimit": 500}
        },
    "color":{"aggregate":"count","type":"quantitative"}
  }
}).add_selection(brush)

rect
```

```python
rect.save(myJekyllSaveDir+'dashboard_plot2_from_dict.json')
```
